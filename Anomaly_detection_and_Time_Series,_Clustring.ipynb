1. What is Dimensionality Reduction? Why is it important in machine
learning?       
  - Dimensionality Reduction is the process of reducing the number of input features (variables) while keeping as much useful information as possible.   
  It transforms high-dimensional data into a lower-dimensional space using techniques like PCA, t-SNE, Autoencoders.     
  This is important decause high-dimensional data often contains noise, redundant information, and irrelevant features, which make models slow and less accurate.     
  Reducing dimensions helps simplify data, improve model performance, and make visualization easier.       
  . Reduces Overfitting: Removes noise and irrelevant features - models generalize better.      
  . Improves Model Speed & Efficiency: Fewer features - faster training and prediction.     
  . Handles Curse of Dimensionality: High - dimensional data becomes sparse; reducing dimensions solves this problem.     
  . Enhances Visualization: Allows data to be plotted in 2D or 3D for better understanding.       
  . Improves Storage & Processing: Reduces dataset size - saves memory and computation.     

2. Name and briefly describe three common dimensionality reduction
techniques.        
  - . Principal Component Analysis (PCA):

  These components capture the maximum variance in the data.

  A non-linear technique mainly used for visualizing high-dimensional data in 2D or 3D.      
  Preserves local structure (similar points stay close).      
  Great for clustering visualization but not good for large datasets.     
  . Autoencoders:     
  Neural networkâ€“based dimensionality reduction.     
  Compress input data into a smaller latent representation and then reconstruct it.      
  Useful for non-linear data, anomaly detection, and feature learning.

3. What is clustering in unsupervised learning? Mention three popular
clustering algorithms.     
  - Clustering is an unsupervised learning technique that groups data points into clusters such that:     
  Points within the same cluster are more similar to each other.      
  And different clusters are as dissimilar as possible.      
  It helps discover hidden patterns, natural groupings, and structure in unlabeled data.     
  Three Popular Clustering Algorithms:      
  K-Means Clustering:      
  Partitions data into K clusters by minimizing within-cluster variance.     
  Hierarchical Clustering:      
  Builds a tree-like structure (dendrogram) using agglomerative or divisive methods.      
  DBSCAN (Density-Based Spatial Clustering of Applications with Noise):     
  Forms clusters based on density of data points and can detect outliers.

4. Explain the concept of anomaly detection and its significance.      
  - Anomaly detection is the process of identifying data points, events, or observations that deviate significantly from the normal pattern of a dataset. These unusual points are known as anomalies or outliers.     
  It is widely used when the goal is to detect rare, unexpected, or suspicious behavior.     
  Significance of Anomaly Detection:       
  Fraud Detection: Helps identify unusual transactions in banking or credit card data.     
  Cybersecurity: Detects abnormal network activity such as intrusions or malware behavior.      
  Fault Detection: Used in manufacturing or IoT systems to detect equipment failures early.    
  Healthcare Monitoring: Identifies abnormal vital signs or unusual patient health trends.     
  Data Cleaning: Helps remove noisy or incorrect records before training machine learning models.

5.  List and briefly describe three types of anomaly detection techniques.       
  - Statistical Methods:      
  These techniques assume that normal data follows a known statistical distribution (like Gaussian).     
  Any data point that lies far from the expected range (e.g., beyond a threshold or confidence interval) is flagged as an anomaly.    
  Distance-Based Methods:     
  These methods detect anomalies by measuring how far a point is from others.     
  If a point has few close neighbors or lies far from the cluster of normal data, it is considered an anomaly.     
  Example: k-Nearest Neighbors (KNN) anomaly detection.     
  Density-Based Methods:      
  These methods assume anomalies occur in low-density regions of the data.      
  If a point is located where the density of nearby points is low, it is labeled as an anomaly.     
  Example: Local Outlier Factor (LOF).

6. What is time series analysis? Mention two key components of time series
data.      
  - Time series analysis is the process of studying data points collected over time to identify patterns, trends, and dependencies.     
  It helps in forecasting future values, understanding seasonal behavior, and detecting changes over time.        
  Two Key Components of Time Series Data:     Trend: A long-term upward or downward movement in the data over time.
Example: increasing sales every year.     
Seasonality: A repeating pattern at regular intervals (daily, monthly, yearly).     
Example: high ice-cream sales every summer.

7. Describe the difference between seasonality and cyclic behavior in time
series.      
  - Seasonality:     
  Definition: Repeating patterns that occur at fixed, regular intervals (daily, weekly, monthly, yearly).     
  Cause: Usually due to calendar-based or environmental factors.    
  Predictability: Highly predictable because the pattern repeats in a consistent period.    
  Example: Higher electricity usage every summer due to AC.     
  Cyclic Behavior:     
  Definition: Long-term up-and-down fluctuations that occur over irregular, variable intervals.      
  Cause: Often influenced by economic or external factors.      
  Predictability: Not fixed or regular; cycles can be long or short.     
  Example: Economic recession followed by recovery (no fixed duration).

8. Write Python code to perform K-means clustering on a sample dataset.      


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y_true = make_blobs(n_samples = 300, centers = 3, cluster_std = 1.0, random_state = 42)

kmeans = KMeans(n_clusters=3, random_state = 42)
kmeans.fit(X)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print("Cluster Centers:\n", centroids)
print("\nFirst 10 Predicted Cluster Labels:\n", labels[:10])

plt.figure(figsize= (8,5))
plt.scatter(X[:,0], X[:,1], c = labels, s = 40)
plt.scatter(centroids[:,0], centroids[:,1], c = "red", s = 200, marker = "X")
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

9. What is inheritance in OOP? Provide a simple example in Python.      
  - Inheritance in Object-Oriented Programming (OOP) is a mechanism where a child class (subclass) can acquire the properties and methods of a parent class (superclass).      
  It helps in code reusability, reducing duplication, and creating a logical class hierarchy.

```python
class Animal:
  def sound(self):
    return "Animals make sounds"

class Dog(Animal):
  def bark(self):
    return "Dog barks"

d = Dog()
print(d.sound())
print(d.bark())
```

  If an actual value is too far from the expected value, it is flagged as an anomaly.    
  . Model learns normal behavior: Time series models (ARIMA, LSTM, Moving Average) learn trends, seasonality, cycles.    
  . Forecast or expected range is generated: For each future point, the model predicts expected values and confidence intervals.     
  . Compare actual vs. expected values: If the actual data point deviates significantly (e.g., beyond 3Ïƒ or outside prediction interval), it is marked as an anomaly.    
  Useful for real-time monitoring: Systems like credit card fraud detection, sensor failure detection, network intrusion use time-based anomaly detection.
